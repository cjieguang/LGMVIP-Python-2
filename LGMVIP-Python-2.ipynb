```python
# Imports
import turtle
import random
import time

score = 0
highscore = 0
delay = 0.1
```

```python
# Initialize window
screen = turtle.Screen()
# Set game title
screen.title("Jie Guang's Snake Game")
# Set screen colour
screen.bgcolor("#C1E1C1")
# Set screen delay
screen.tracer(0)
# Set screen size
screen.setup(width = 620, height = 620)

# Create snake
snakeHead = turtle.Turtle(shape = "square")
snakeHead.color("red")
snakeHead.penup()
snakeHead.goto(0, 0)
snakeHead.direction = "Stop"

# Making Food
food = turtle.Turtle()
foodColour = random.choice(['blue', 'green', 'yellow'])
foodShape = random.choice(['triangle', 'circle', 'square'])
food.speed(0)
food.shape(foodShape)
food.color(foodColour)
food.penup()
food.goto(0, 150)

# Making scoreboard
pen = turtle.Turtle()
pen.speed(0)
pen.shape("circle")
pen.color("black")
pen.penup()
pen.hideturtle()
pen.goto(-300, 240)
pen.write("Score: 0\nHigh Score: 0", align = "left", font = ("arial", 14, "bold"))



# Define Key Press
# Cannot go direct opposite of current direction, need to turn, else it'll hit itself
def goUp():
    if snakeHead.direction != "down":
        snakeHead.direction = "up"
def goDown():
    if snakeHead.direction != "up":
        snakeHead.direction = "down"
def goLeft():
    if snakeHead.direction != "right":
        snakeHead.direction = "left"
def goRight():
    if snakeHead.direction != "left":
        snakeHead.direction = "right"
def move():
    if snakeHead.direction == "up":
        y_pos = snakeHead.ycor()
        snakeHead.sety(y_pos+20)
    if snakeHead.direction == "down":
        y_pos = snakeHead.ycor()
        snakeHead.sety(y_pos-20)
    if snakeHead.direction == "left":
        x_pos = snakeHead.xcor()
        snakeHead.setx(x_pos-20)
    if snakeHead.direction == "right":
        x_pos = snakeHead.xcor()
        snakeHead.setx(x_pos+20)

screen.listen()
screen.onkeypress(goUp, "w")
screen.onkeypress(goLeft, "a")
screen.onkeypress(goDown, "s")
screen.onkeypress(goRight, "d")


bodies = []

# Main loop
while True:
    pen.clear()
    pen.write(f"Score: {score}\nHigh Score: {highscore} ", align = "left", font = ("arial", 14, "bold"))
    screen.update()
    # Check if snake hits border
    if snakeHead.xcor() > 300 or snakeHead.ycor() > 300 or snakeHead.xcor() < -300 or snakeHead.ycor() < -300:
        snakeHead.goto(0, 0)
        snakeHead.direction = "Stop"

        # Clear dead bodies
        for body in bodies:
            body.goto(1200, 1200)
        bodies.clear()

        # Update scoreboard
        score = 0
        pen.clear()
        pen.write(f"Score: {score}\nHigh Score: {highscore}\nGAME OVER", align = "left", font = ("arial", 14, "bold"))
        time.sleep(2)
        continue

    # Check if fruit was eaten
    if snakeHead.distance(food) < 20:

        # use random range to input steps of 20, ensure fruit is on the grid
        x_pos = random.randrange(-260, 280, 20)
        y_pos = random.randrange(-260, 280, 20)

        # Randomize fruits again
        foodColour = random.choice(['blue', 'green', 'yellow'])
        foodShape = random.choice(['triangle', 'circle', 'square'])
        food.shape(foodShape)
        food.color(foodColour)

        # Change fruit location
        food.goto(x_pos, y_pos)

        # Adding body
        newBody = turtle.Turtle()
        newBody.speed(0)
        newBody.shape("square")
        newBody.color("red")
        newBody.penup()
        bodies.append(newBody)
        
        # Highscore check
        score += 10
        if score > highscore:
            highscore = score

        # Update scoreboard
        pen.clear()
        pen.write(f"Score: {score}\nHigh Score: {highscore} ", align = "left", font = ("arial", 14, "bold"))


    # Visual updates for movement
    for index in range(len(bodies)-1, 0, -1):
        # Shifts bodies to the front body's location
        x_pos = bodies[index-1].xcor()
        y_pos = bodies[index-1].ycor()
        bodies[index].goto(x_pos, y_pos)
    
    # Shifts front most body to head's location
    if len(bodies) > 0:
        x_pos = snakeHead.xcor()
        y_pos = snakeHead.ycor()
        bodies[0].goto(x_pos, y_pos)
    
    # Update of head movement
    move()

    # Checks for snake's head collision
    for body in bodies:
        if body.distance(snakeHead) < 20:
            snakeHead.goto(0, 0)
            snakeHead.direction = "stop"

            # Clear dead bodies
            for body in bodies:
                body.goto(1200, 1200)
            bodies.clear()

            # Update scoreboard
            score = 0
            pen.clear()
            pen.write(f"Score: {score}\nHigh Score: {highscore}\nGAME OVER", align = "left", font = ("arial", 14, "bold"))
            time.sleep(2)
    time.sleep(delay)
screen.mainloop()    
```
